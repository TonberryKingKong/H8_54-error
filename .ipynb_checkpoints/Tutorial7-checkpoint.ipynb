```python
df_can = pd.read_excel('https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true',
             sheet_name = 'Canada by Citizenship',
             skiprows=range(20),
             skipfooter=2)

print('Data read into pandas DataFrame!')
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
Japan = df_can.loc['Japan', years]
Japan. head()
```

```python
Japan.index = Japan.index.map(int)
Japan.plot(kind='line')
plt.title('Immigration from Japan')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
Haiti_Japan = df_can.loc[['Haiti','Japan'],years]
Haiti_Japan
```

```python
Haiti_Japan.plot(kind='line')
```

```python
df_HJ=Haiti_Japan.transpose()
df_HJ.head()
```

```python
df_HJ.plot(kind='line')
plt.title('Comparison Haiti VS Japan')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_can.sort_values(by='Total',ascending=False,axis=0,inplace=True)
df_Top5 = df_can.head(5)
df_Top5 = df_Top5[years].transpose()
df_Top5.index = df_Top5.index.map(int)
df_Top5.plot(kind='line',figsize= (14,8))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_can.sort_values(['Total'],ascending=False, axis =0, inplace=True )
df_Top5 = df_can.head()
df_Top5 = df_Top5[years].transpose()
df_Top5.head()
```

```python
df_Top5.plot(kind='area',
            stacked=False,
            figsize=(20,10),
            )
plt.title('Immigration trends of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
ax = df_Top5.plot(kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10),
            )
plt.title('Immigration trends of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
ax.set_title('Immigration Trend of Top 5 countries')
ax.set_ylabel('Number of immigrants')
ax.set_xlabel('Years')
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()
```

```python
df_least5.plot(kind='area',
              alpha=0.45,
              stacked=True,
              figsize=(20,10),
              )
plt.title('Immigration Trend of least 5 countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
