```python
df1 = pd.read_csv(r'https://ww2.amstat.org/censusatschool/Exports/1618806301247_DATA.csv')
df1.head()
```

```python
df1.info()
```

```python
df1 = df1.dropna(subset=['Height_cm'])
```

```python
def convert_feet_to_cm(x):
    if '.' in x:
        return int(float(x))
    if "'" not in x and ";" not in x:
        return int(x)
    if "'" in x:
        feet_inch =x.split("'")
    elif ";" in x:
        feet_inch =x.split(";")
        feet = int(feet_inch[0])
        inch = int(feet_inch[1])
        height = 0
        height += feet * 30.48
        height += inch *2.54
        return int(height)
```

```python
df1.describe()
```

```python
df1['Height_cm2'] = df1['Height_cm'].apply(convert_feet_to_cm)
df1['Height_cm2']
```

```python
df1[df1['Height_cm2']>300]
```

```python
height_mean = df1['Height_cm2'].mean()
height_mean
```

```python
height_std = df1['Height_cm2'].std()
height_std
```

```python
z_score = 1.96
lcb = height_mean - z_score * height_std
ucb = height_mean + z_score * height_std
lcb,ucb
```
